#### defining functions

```python
# defining the condition
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# applying on df
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# check output
df['price_range'].value_counts(dropna = False)
```

```python
# max value
df['prices'].max()
```

#### IF LOC()

```python
# IF with LOC()
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# check output
df['price_range_loc'].value_counts(dropna = False)
```

```python
# IF with LOC() on entire df
df_orders_products_merged.loc[df_orders_products_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_orders_products_merged.loc[(df_orders_products_merged['prices'] <= 15) & (df_orders_products_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

#### IF with FOR-loops

```python
# frequency of a orders dow column
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# creating a new column
result = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

## 4.7.4 Create “busiest_period_of_day” column

```python
# frequency of a orders dow column
df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# create new column
result_3 = []

for value in df_orders_products_merged["order_hour_of_day"]:
  if value == 9 or value == 10 or value == 11 or value == 12 or value == 13 or value == 14 or value == 15 or value == 16:
    result_3.append("Most orders")
  elif value == 0 or value == 1 or value == 2 or value == 3 or value == 4 or value == 5 or value == 6:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
# print the result
result_3
```

```python
# create new column for the results
df_orders_products_merged['busiest_period_of_day'] = result_3
```
