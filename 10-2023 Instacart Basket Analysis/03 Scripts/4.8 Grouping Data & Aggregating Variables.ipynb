## 4.8.6 Spending flag

```python
# create new column
ords_prods_merge['mean_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check
ords_prods_merge[['user_id', 'mean_spend']].head(10)
```

```python
# create flag column
ords_prods_merge.loc[ords_prods_merge['mean_spend'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend'] >= 10, 'spender_flag'] = 'High spender'
```

```python
#check output
ords_prods_merge[['user_id', 'mean_spend', 'spender_flag']].head(30)
```

```python
# check with frequency
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

## 4.8.7 Order frequency flag

```python
# create new column
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check
ords_prods_merge[['user_id', 'median_days_since_prior_order']].head(10)
```

```python
# create flag column
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#check output
ords_prods_merge[['user_id', 'median_days_since_prior_order', 'order_frequency_flag']].head(30)
```

```python
# check with frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# NaN check
df_check = ords_prods_merge[ords_prods_merge['order_frequency_flag'].isnull() == True]
```

```python
df_check
```
